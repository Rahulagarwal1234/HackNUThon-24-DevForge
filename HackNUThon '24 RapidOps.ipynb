```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import re

data = pd.read_csv("/content/fashion-dataset/styles.csv", nrows=6000)
stemmer = PorterStemmer()

for i in data.columns:
    if data[i].dtype == 'O':  # Check if the column is of object (string) type
        data[i] = data[i].str.lower()  # Convert the text to lowercase

        # Preprocess text to remove non-alphanumeric characters
        data[i] = data[i].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))

        # Tokenize, stem, and rejoin the words
        data[i] = data[i].apply(lambda x: ' '.join([stemmer.stem(word) for word in word_tokenize(x)]))

print(data['articleType'].unique())
```

```python
keyword_map=dict()
for column_name in data.columns:
    if data[column_name].dtype == 'object':
        unique_values = data[column_name].unique()
        for value in unique_values:
            key = f"{column_name}={value}"
            if key not in keyword_map:  # Ensure keys are unique
                keyword_map[value] = key

    print()
print(keyword_map)
```
